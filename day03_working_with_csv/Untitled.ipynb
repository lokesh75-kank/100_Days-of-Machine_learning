## 2. Openging a csv file with url/ or from server directly


```python
import requests
from io import StringIO

url = "https://raw.githubusercontent.com/cs109/2014_data/master/countries.csv"
headers = {"User-Agent": "Mozilla/5.0 (Macintosh; Intel Mac OS X 10.14; rv:66.0) Gecko/20100101 Firefox/66.0"}
req = requests.get(url , headers = headers)
file = StringIO(req.text)
df = pd.read_csv(file)
```

## 3. Sep Parameter

```python
#let import tsv file and change the default sep parameter to read the file and also give the column 
#names or else it will take first row as the col names.

df = pd.read_csv("movie_titles_metadata.tsv", sep = "\t", names=['sno','name','release_year','rating','votes','genres'])
df.head()
```

## 4. Index_col parameter

```python
#this paramter is used when you want to give your column as a index
df = pd.read_csv("aug_train.csv",index_col = "enrollee_id")
df.head()
```

## 5. Header parameter

```python
#sometime in file columns become row and stats the data from their, so to fix this use header parameter
df = pd.read_csv("test.csv",header= 1)
df.head()
```

## 6. use_cols parameter

```python
#when we need specific columns only from the file then we can use this parameter
df = pd.read_csv("aug_train.csv", usecols=['enrollee_id','gender','education_level'])
df.head()
```

## 7. Squeeze parameters

```python
#this parameter is use to reture the df in series format
df = pd.read_csv("aug_train.csv", squeeze= True)
df.head()
```

## 8. Skiprows/nrows Parameter

```python
# Skiprows is used to skip the number of rows from the data set and nrow give the required number of rows
#index for skiprow starts from 1.
df = pd.read_csv("aug_train.csv",skiprows=[3,5], nrows= 100)
df
```

## 9. Encoding Parameter

```python
#Sometimes the csv file dont have default UTF8 encoding parameter so in this case we have to specify the encoding,
#To determine the encoding of a CSV file, you can use the chardet library in Python. chardet is a Python library that detects 
#the encoding of a file by analyzing the byte sequences.

import chardet

with open("zomato.csv","rb") as f:
    result = chardet.detect(f.read())
print("encoding",result)
```

```python
#Open the file in a text editor that allows you to manually set the encoding. Look at the file contents and see if it makes sense. 
#If it does, note down the encoding and use it with Pandas.
df = pd.read_csv("zomato.csv", encoding='latin-1')
print(df.head())
```

##  12. Handling dates in csv file

```python
df = pd.read_csv('IPL Matches 2008-2020.csv')
# df.info()
#we can see the date parameter has object as data type , so need to change it to datattime type
df = pd.read_csv('IPL Matches 2008-2020.csv', parse_dates=['date'])
df.info()
```

## 13. Convertors: it is used to convert the data in predictor.

```python

def rename(name):
    if name == "Royal Challengers Bangalore":
        return "RCB"
    else:
        return name
```

```python
df = pd.read_csv('IPL Matches 2008-2020.csv')
df['team1'].head()
```

```python
#using convertor when loading the data

df = pd.read_csv('IPL Matches 2008-2020.csv',converters={"team1":rename})

df['team1'].head()
```

## 14. na_values Parameter

```python
# its used to replace any values in col with NaN 
#lets replace bangalore to NaN
df = pd.read_csv('IPL Matches 2008-2020.csv',na_values= ['Bangalore'])
df.head()
#lets replace bangalore to NaN
```
